```python
#Dependencies
import pandas as pd
import numpy as np
```

```python
#File to load
pymoli_file = "Resources/purchase_data.csv"
```

```python
#Read file
purchase_data = pd.read_csv(pymoli_file)
pd.read_csv("Resources/purchase_data.csv")
purchase_data.head(10)
```

```python
purchase_data.describe().round(2)
```

```python
# Calculate the Number of Unique Players 
player_count = len(purchase_data["SN"].unique())

pd.DataFrame({"Total Players": [player_count]})
```

```python
#Average purchasing price 
average_price = purchase_data["Price"].mean()
```

```python
#Total number of purchases 
total_purchases = purchase_data["Item ID"].count()
```

```python
#Total of unique items 
unique_items = len(purchase_data["Item Name"].unique())

#Total revenue
total_revenue = purchase_data["Price"].sum()

summary_df= pd.DataFrame({"Unique Items": unique_items, "Average Price ($)": average_price, "Total Purchases": total_purchases, "Total Revenue ($)": [total_revenue]})

summary_df.round(2)
```

```python
#Percentage and Count of Male Players
#Percentage and Count of Female Players
#Percentage and Count of Other / Non-Disclosed


gender_data = purchase_data.loc[:,[ "SN", "Gender","Age"]]
gender_data = gender_data.drop_duplicates()


gender_count = gender_data["Gender"].value_counts()
gender_count

percentages_gender = gender_count/player_count * 100
percentages_gender

summary_gender = pd.DataFrame({"Total Count": gender_count,"Percentage of Players": percentages_gender })
summary_gender.round(2)
```

```python
#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
#Create a summary data frame to hold the results
#Give the displayed data cleaner formatting
#Display the summary data frame 

mean_gender = purchase_data.groupby(["Gender"]).mean()["Price"]
mean_gender

count_gender = purchase_data.groupby(["Gender"]).count()["Price"]
count_gender

total_gender = purchase_data.groupby(["Gender"]).sum()["Price"]
total_gender

per_person = total_gender/summary_gender["Total Count"]

summary_totals = pd.DataFrame({"Purchase Count": count_gender, "Average Purchase Price": mean_gender, "Total Purchase Value": total_gender, "Average Total Purchase per Person": per_person})
summary_totals.round(2)
```

```python
#Age Demographics
#Establish bins for ages
#Categorize the existing players using the age bins
#Calculate the numbers and percentages by age group
#Create a summary data frame to hold the results
#Round the percentage column to two decimal points
#Display Age Demographics Table

age_bins = [0, 9.9, 14.9, 19.90, 24.90, 29.90, 34.90, 39.90, 9999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Categorize the existing players using the age bins

purchase_data["Age Ranges"]= pd.cut(purchase_data["Age"], age_bins, labels=group_names)

age_count = purchase_data["Age Ranges"].value_counts()


age_demographics_percents = age_count / player_count * 100


age_demographics = pd.DataFrame({"Total Count": age_count, "Percentage of Players": age_demographics_percents})

age_demographics

# Minor Data Munging
age_demographics = age_demographics.round(2)

# Display Age Demographics Table
age_demographics.sort_index()
```

```python
#Purchasing Analysis (Age) -
#Bin the purchase_data data frame by age
##Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
#Create a summary data frame to hold the results
#Optional: give the displayed data cleaner formatting
#Display the summary data frame

age_purchase_total = purchase_data.groupby(["Age Ranges"]).sum()["Price"].rename("Total Purchase Value")

age_purchase_count = purchase_data.groupby(["Age Ranges"]).count()["Price"].rename("Purchase Count")

av_purchase_value = purchase_data.groupby(["Age Ranges"]).mean()["Price"].rename("Average Purchase Price")

av_per_person = age_purchase_total / age_purchase_count

age_purchase_total_summary = pd.DataFrame({"Purchase Count": age_purchase_count, 
                                           "Total Purchase Value": age_purchase_total, 
                                           "Average Purchase Price": av_purchase_value, 
                                           "Average Per Person": av_per_person})
age_purchase_total_summary.round(2)
```

```python
#Top Spenders 
#Identify the the top 5 spenders in the game by total purchase value, then list (in a table):
#SN
#Purchase Count
#Average Purchase Price
#Total Purchase Value

unique_purchase = purchase_data["SN"].value_counts()

total_purchase = purchase_data.groupby("SN")["Price"].sum()

av_purchase = total_purchase/unique_purchase

summary = pd.DataFrame({"Purchase Count": unique_purchase, "Total Purchase": total_purchase, "Average Purchase": av_purchase})

largest = summary.sort_values("Total Purchase", ascending=False).round(2)

largest.head(5)
```

```python
#Most Popular Items 
#Identify the 5 most popular items by purchase count, then list (in a table):
#Item ID
#Item Name
#Purchase Count
#Item Price
#Total Purchase Value

most_popular_items = purchase_data["Item Name"].value_counts()
most_popular_items

total_purchaseIN = purchase_data.groupby("Item Name")["Price"].sum()
total_purchaseIN

final_items = total_purchaseIN/most_popular_items

summary2 = pd.DataFrame({"Most Popular": most_popular_items, "Total Purchases":total_purchaseIN, "Item Price": final_items})

largest = summary2.sort_values("Total Purchases", ascending=False).round(2)

largest.head(5)
```
